```python
import os
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
import shutil

path = os.path.join(os.getcwd(), "datasets")
subdir = os.listdir(path)
Imagepath = os.path.join(path, subdir[0])
labelpath = os.path.join(path, subdir[1])

files = [file for file in os.listdir(Imagepath) if (file.find("56845") != -1 or file.find("57634") != -1 )]

filesfile = pd.DataFrame(files)
train, val, test = np.split(filesfile.sample(frac=1, random_state=42), [int(.6*len(filesfile)), int(.8*len(filesfile))])
print(len(train), len(val), len(test))
train, val, test

train_imgpath = os.path.join(Imagepath, "train")
val_imgpath = os.path.join(Imagepath, "val")
test_imgpath = os.path.join(Imagepath, "test")

train_labelpath = os.path.join(labelpath, "train")
val_labelpath = os.path.join(labelpath, "val")
test_labelpath = os.path.join(labelpath, "test")
for file in train[0]:
    try:
        print(file)
        img = cv2.imread(os.path.join(Imagepath, file))
        cv2.imwrite(os.path.join(train_imgpath, file), img)
        

        shutil.copyfile(os.path.join(labelpath, file[:-3]+"txt"), os.path.join(train_labelpath, file[:-3]+"txt"))
    except:
        pass
for file in val[0]:
    try:
        print(file)
        img = cv2.imread(os.path.join(Imagepath, file))
        cv2.imwrite(os.path.join(val_imgpath, file), img)
        

        shutil.copyfile(os.path.join(labelpath, file[:-3]+"txt"), os.path.join(val_labelpath, file[:-3]+"txt"))
    except:
        pass
for file in test[0]:
    try:
        print(file)
        img = cv2.imread(os.path.join(Imagepath, file))
        cv2.imwrite(os.path.join(test_imgpath, file), img)
        

        shutil.copyfile(os.path.join(labelpath, file[:-3]+"txt"), os.path.join(test_labelpath, file[:-3]+"txt"))
    except:
        pass
    
```
